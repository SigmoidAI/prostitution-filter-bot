```python
# Adding a new data to our data for negative samples.
# The data set: https://www.kaggle.com/uciml/sms-spam-collection-dataset
df = pd.read_csv('spam.csv', encoding='latin-1')
df
```

```python
# Dropping out the unsefull columns.
df = df.drop(['Unnamed: 2', 'Unnamed: 3', "Unnamed: 4"], axis=1)
```

```python
# Keeping only the 'ham' samples.
df = df[df['v1']=='ham']
```

```python
df
```

```python
bad_txt
```

```python
# Keeping all documents except the first.
bad_txt = bad_txt[1:]
```

```python
# Generating the positive label list.
bad_label = np.ones(len(bad_txt))
```

```python
bad_label
```

```python
# Generating the negative label list.
good_label = np.zeros(len(df))
```

```python
# Concatenating the label lists.
labels = np.concatenate([bad_label, good_label])
```

```python
# Extracting the text for negative samples from data frame.
good_txt = df['v2'].values
```

```python
# Concatenating the poasitive and negative textes.
txt = np.concatenate([bad_txt, good_txt])
```

```python
# Lovering alll texts.
txt = [text.lower() for text in txt]
```

```python
# Keeping only the alphanumeric values.
import re
txt = [' '.join(re.findall('[a-zA-Z]+', text)) for text in txt]
txt
```

```python
# Keeping onlu the word that have more than one symbol.
for i in range(len(txt)):
    txt[i] = ' '.join([word for word in txt[i].split() if len(word) >= 2])
txt
```

```python
# Importing the nltk modules.
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
# Removing the stopwords from documents.
for i in range(len(txt)):
    txt[i] = ' '.join([word for word in txt[i].split() if not word in stopwords.words('english')])
txt
```

```python
# Stemming all documents.
stemmer = PorterStemmer()
for i in range(len(txt)):
    txt[i] = ' '.join([stemmer.stem(word) for word in txt[i].split()])
txt
```

```python
# Importing the TF-IDF Vectroizer from sklearn and Frequency Distribution from nltk.
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk import FreqDist
```

```python
# Cretting the frequncy distribution for the corpus.
from nltk.tokenize import word_tokenize
corpus = FreqDist()
for i in range(len(txt)):
    for word in word_tokenize(txt[i]):
        corpus[word] +=1
```

```python
# There are no hapaxes.
len(corpus.hapaxes())
```

```python
for i in range(len(txt)):
    txt[i] = ' '.join([word for word in txt[i].split() if word not in corpus.hapaxes()])
txt
```

```python
len(corpus)
```

```python
# Vectorizing the text.
vectorizer = TfidfVectorizer(min_df=5)
X = vectorizer.fit_transform(txt)
```

```python
X.toarray().shape
```

```python
# Importing the testing functions and the Gaussian Naive Bayes model
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.naive_bayes import GaussianNB
gauss = GaussianNB()
```

```python
# Spliting the data into train and test set.
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, shuffle=True)
```

```python
y_test
```

```python
# Gettong the cross validation ofthe model.
cross_val_score(gauss, X_train.todense(), y_train)
```

```python
# Fitting the model.
gauss.fit(X_train.todense(), y_train)
```

```python
# Generating the predictions.
yhat = gauss.predict(X_test.todense())
```

```python
# Looking at hte accuracy socre.
from sklearn.metrics import accuracy_score
accuracy_score(y_test, yhat)
```

```python
# Creatting the confusion matrix.
from sklearn.metrics import confusion_matrix
confusion_matrix(yhat, y_test.reshape(-1, 1))
```

```python
# Ploting the confusion matrix.
sns.heatmap(confusion_matrix(yhat, y_test.reshape(-1, 1),normalize=True), annot=True)
```

```python
# Creating the Text Normalizer model.
# take a look in the norm.py module for a better version of it.
import re
class TextNormalizer():
    def __init__(self, language = 'english'):
        self.stopwors = stopwords.words(language)
        self.stemmer = PorterStemmer()
        self.vocabulary = list(dict(corpus).keys())
    def transform(self, X):
        for i in range(len(X)):
            X[i] = X[i].replace('\n', ' ')
            X[i] = X[i].replace('\r', ' ')
            X[i] = re.sub('[^\x00-\x7F]+', '', X[i])
            X[i] = X[i].lower()
            X[i] = ' '.join(re.findall('[a-zA-Z]+', X[i]))
            X[i] = ' '.join([word for word in X[i].split() if len(word) >= 2])
            X[i] = ' '.join([word for word in X[i].split() if not word in self.stopwors])
            X[i] = ' '.join([self.stemmer.stem(word) for word in X[i].split()])
            X[i] = ' '.join([word for word in X[i].split() if word in self.vocabulary])
        return X
```
